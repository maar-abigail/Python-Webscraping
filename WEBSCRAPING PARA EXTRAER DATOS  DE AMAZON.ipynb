```python
#USANDO WEBSCRAPING PARA EXTRAER DATOS 

#Instalamos todas las paqueterias necesarias
!pip install pandas==1.3.3
!pip install requests==2.26.0
!mamba install bs4==4.10.0 -y
!mamba install html5lib==1.1 -y 
!pip install lxml==4.6.4
!pip install plotly==5.3.1

import pandas as pd
import requests
from bs4 import BeautifulSoup

#Ocupamos la biblioteca Request para enviar una solicitud a HTTP a la pagina web
url =  "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/amazon_data_webpage.html" 

#Utilizamos request.get() para especificar la ubicacion de donde queremos recuperar la informacion y .text para extraer el contenido HTML como una cadena legible
data  = requests.get(url).text
print(data)

#Para analizar los datos se llama a la biblioteca BeautifulSoup
soup = BeautifulSoup(data, 'html5lib')

#Para una mejor visualizacion se creara un DataFrame vacio usando la funcion pd.DataFrame con algunos datos relevantes.
amazon_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close","Volume"])

#Ahora extraemos los datos mediante BeautifulSoup con la siguiente informacion
#Estas son las siguientes etiquetas que se utilizan al crear tablas HTML.
#Etiqueta <table>: esta etiqueta es la etiqueta raíz que se utiliza para definir el inicio y el final de la tabla. Todo el contenido de la tabla está incluido dentro de estas etiquetas.
#Etiqueta <tr>: esta etiqueta se utiliza para definir una fila de la tabla. Cada fila de la tabla se define dentro de esta etiqueta.
#Etiqueta <td>: esta etiqueta se utiliza para definir una celda de la tabla. Cada celda de la tabla se define dentro de esta etiqueta. Puede especificar el contenido de la celda entre las etiquetas de apertura y cierre.
#Etiqueta <th>: esta etiqueta se utiliza para definir una celda de encabezado en la tabla. La celda de encabezado se utiliza para describir el contenido de una columna o fila. De forma predeterminada, el texto dentro de una etiqueta está en negrita y centrado.
#Etiqueta <tbody>: este es el contenido principal de la tabla, que se define mediante la etiqueta. Contiene una o más filas de elementos.
amazon_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])

for row in soup.find("tbody").find_all("tr"):
    col = row.find_all("td")
    date =col[0].text
    Open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text

#Agregamos los datos de cada fila a la tabla 
amazon_data = amazon_data.append({"Date":date, "Open":Open, "High":high, "Low":low, "Close":close, "Adj Close":adj_close, "Volume":volume}, ignore_index=True)

#Ahora llenamos el DataFrame con la informacion encontrada y la funcion head() o tail()
amazon_data.head()
```
